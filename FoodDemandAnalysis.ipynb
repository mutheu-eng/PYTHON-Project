DESCRIPTIVE ANALYSIS : 
UNIVARIATE ANALYSIS

```python
#sum of base prices
Total_BasePrices = data['base_price'].sum()
formatted_total = "{:,.2f}".format(Total_BasePrices)

print("Total Base Prices:", formatted_total)
```

```python
#sum of num of orders
Total_Num_orders = data['num_orders'].sum()
formatted_total = "{:,.2f}".format(Total_Num_orders)

print("Total Number of Orders:", formatted_total)
```

```python
#sum of checkout prices
Total_CheckoutPrices = data['checkout_price'].sum()
formatted_total = "{:,.2f}".format(Total_CheckoutPrices)

print("Total Checkout Prices:", formatted_total)
```

```python
Average_BasePrices = data['base_price'].mean()
formatted_mean = "{:,.2f}".format(Average_BasePrices)

print("Mean of Base Prices:", formatted_mean)
```

```python
Average_Num_orders = data['num_orders'].mean()
formatted_mean = "{:,.2f}".format(Average_Num_orders)

print("Mean Number of Orders:", formatted_mean)
```

```python
Average_CheckoutPrices = data['checkout_price'].mean()
formatted_mean = "{:,.2f}".format(Average_CheckoutPrices)

print("Mean Checkout Prices:", formatted_mean)
```

CHECKING OUT OYLIER IN OUR CHECK OUT PRICES

```python
sns.boxplot(x=data['checkout_price'])
plt.title('BoxPlot of Check Out Prices')
plt.xlabel('CheckOut Price')
plt.show()
```

BIVARIATE ANALYSIS

```python
category_sum_orders = data.groupby('category')['num_orders'].sum().sort_values(ascending=False)

plt.figure(figsize=(7,5))
category_sum_orders.plot(kind='bar')
plt.title('Total Orders by Category')
plt.xlabel('Category')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
category_sum_orders = data.groupby('cuisine')['num_orders'].sum().sort_values(ascending=False)

plt.figure(figsize=(7,5))
category_sum_orders.plot(kind='bar')
plt.title('Total Orders by Category')
plt.xlabel('Cuisine')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
center_type_count = data.groupby('center_type')['num_orders'].count().sort_values(ascending=False)

plt.figure(figsize=(7, 5))
center_type_count.plot(kind='bar')
plt.title('Total Orders by Center Type')
plt.xlabel('Center Type')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
homepage_featured_count = data.groupby('homepage_featured')['num_orders'].count().sort_values(ascending=False)

plt.figure(figsize=(7, 5))
homepage_featured_count.plot(kind='bar')
plt.title('Total Orders by Homepage Featured')
plt.xlabel('Homepage Featured')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
emailer_promotion_count = data.groupby('emailer_for_promotion')['num_orders'].count().sort_values(ascending=False)

plt.figure(figsize=(7, 5))
emailer_promotion_count.plot(kind='bar')
plt.title('Total Orders by Emailer Promotion')
plt.xlabel('Emailer Promotion')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
customer_order_count = data.groupby('id')['num_orders'].count().sort_values(ascending=False)

top_n = 10
top_customers = customer_order_count.head(top_n)

top_customers_df = pd.DataFrame({'Customer ID': top_customers.index, 'Number of Orders': top_customers.values})

print(top_customers_df)
```

```python
customer_order_count = data.groupby('id')['checkout_price'].sum().sort_values(ascending=False)

top_n = 10
top_customers = customer_order_count.head(top_n)

top_customers_df = pd.DataFrame({'Customer ID': top_customers.index, 'Total of Checkout Price': top_customers.values})

print(top_customers_df)
```
